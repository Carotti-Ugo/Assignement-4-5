### Feature Selection
For the feature selection we will use a technique called Variance Inflation Factor (VIF) and analyze the correlation. (For further information about this technique check: https://www.reneshbedre.com/blog/variance-inflation-factor.html) 

```python
# Import the libraries necessaries to do the feature selection
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

#Get all variables except label and id. 
X = df_train[['visitTime','purchaseTime','hour','C1','C2','C3','C4','C5','C6','C7','C8','N1','C9',
              'N2','N3','N4','N5','N6','N7','N8','N9','N10','C10','C11','C12']]   # independent variables
y = df_train['label']   # dependent variables
X.head()
```

```python
# add a constant
X = sm.add_constant(X)

# fit the regression model
reg = sm.OLS(y, X).fit()

# print the VIF results. Above 2.0 means high multicolinearity
pd.DataFrame({'variables':X.columns[1:], 'VIF':[variance_inflation_factor(X.values, i+1) for i in range(len(X.columns[1:]))]})
```

From the VIF analysis we can see that:
  - visitTime
  - hour
  - N3
  - N4
  - N6
  - N8
  - N9
  - N10
  
Have a high multicolinearity, therefore to identify with which features they have high relationship we need to print the correlation matrix, and remove those features that have high relationship with these and run again VIF analysis until we get all of them with an score less than 2.0.

```python
# define X again without the constant
X = df_train[['visitTime','purchaseTime','hour','C1','C2','C3','C4','C5','C6','C7','C8','N1','C9',
              'N2','N3','N4','N5','N6','N7','N8','N9','N10','C10','C11','C12']]   # independent variables

# get the correlations
#X.corr().to_csv("correlations.csv",sep=";")
X.corr()
```

```python
# Remove those that have high correlation hour, N6, N8, N10
X = df_train[['visitTime','purchaseTime','C1','C2','C3','C4','C5','C6','C7','C8','N1','C9',
              'N2','N3','N4','N5','N7','N9','C10','C11','C12']]   # independent variables
X = sm.add_constant(X)

# Running again the VIF
pd.DataFrame({'variables':X.columns[1:], 'VIF':[variance_inflation_factor(X.values, i+1) for i in range(len(X.columns[1:]))]})
```

### Model Development
After selecting the features and dropping those that are not necessary (hour, N6, N8, N10), we can proceed with the most important features of the model.

```python
# Import Libraries
from sklearn.linear_model import LogisticRegression

# Train dataset
X_train = df_train[['visitTime','purchaseTime','C1','C2','C3','C4','C5','C6','C7','C8','N1','C9',
              'N2','N3','N4','N5','N7','N9','C10','C11','C12']]   # independent variables
y_train = y

log_model = LogisticRegression()
log_model.fit(X_train,y_train)
```

### Testing the dataset
In this section we process the test dataset and review the modeling results with the test dataset.

```python
# Import the classification report
from sklearn.metrics import classification_report

# Select the features of the test dataset
X_test = df_test[['visitTime','purchaseTime','C1','C2','C3','C4','C5','C6','C7','C8','N1','C9',
              'N2','N3','N4','N5','N7','N9','C10','C11','C12']]
y_test = df_test['label']

# Make the predictions
predictions = log_model.predict(X_test)

# Print a report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))

# Prediction probabilities to make a purchase (+1).
prob_purchase = log_model.predict_proba(X_test)[:,1]
prob_purchase = pd.DataFrame(data=prob_purchase.flatten())
prob_purchase = prob_purchase.rename(columns = {0:'prob'})
print("Prediction Purchase Probabilities")
prob_purchase
```

Let's try now with the algorithm called SMOTE. 
The SMOTE technique generates new instances of the minority class by interpolating the values of the closest minority instances to a given one.

```python
#We need first to install the package imblearn in order to import the SMOTE module
#!pip install imblearn
from imblearn.over_sampling import SMOTE
```

```python
print("Before OverSampling, counts of label '1': {}".format(sum(y_train == 1)))
print("Before OverSampling, counts of label '-1': {} \n".format(sum(y_train == -1)))

sm = SMOTE(random_state = 2)
X_train_smote, y_train_smote = sm.fit_resample(X_train, y_train.ravel())

print('After OverSampling, the shape of X_train: {}'.format(X_train_smote.shape))
print('After OverSampling, the shape of y_train: {} \n'.format(y_train_smote.shape))

print("After OverSampling, counts of label '1': {}".format(sum(y_train_smote == 1)))
print("After OverSampling, counts of label '-1': {}".format(sum(y_train_smote == -1)))
```

We are also going to try NearMiss algorithm. It refers to a collection of undersampling methods that select examples based on the distance of majority class examples to minority class examples

```python
from imblearn.under_sampling import NearMiss
```

```python
print("Before Undersampling, counts of label '1': {}".format(sum(y_train == 1)))
print("Before Undersampling, counts of label '-1': {} \n".format(sum(y_train == -1)))

nr = NearMiss()

X_train_nearmiss, y_train_nearmiss = nr.fit_resample(X_train, y_train.ravel())

print('After Undersampling, the shape of X_train: {}'.format(X_train_nearmiss.shape))
print('After Undersampling, the shape of y_train: {} \n'.format(y_train_nearmiss.shape))

print("After Undersampling, counts of label '1': {}".format(sum(y_train_nearmiss == 1)))
print("After Undersampling, counts of label '-1': {}".format(sum(y_train_nearmiss == -1)))
```

## Testing of the model with the dataset balanced

```python
log_model2 = LogisticRegression()
log_model2.fit(X_train_nearmiss, y_train_nearmiss.ravel())
predictions_nearmiss = log_model2.predict(X_test)


# print classification report
print(classification_report(y_test, predictions_nearmiss))
```
